```python
import os

import nltk
import pandas as pd
import numpy as np

from pathlib import Path
from ydata_profiling import ProfileReport
from nltk.corpus import stopwords
```

```python
project_dir = os.getcwd().split('notebooks')[0]
```

```python
processed_path = os.path.join(project_dir, 'data', 'processed', 'Steam Games 2024.csv')
```

```python
df = pd.read_csv(processed_path)
```

```python
columns_to_scale = [
    'AppID', # NOT SCALLABLE
    'Estimated owners', 'Peak CCU', # Integers
    'Positive', 'Negative', 'Score rank', # Integers
    'Achievements', 'Recommendations', # Integers 
    'Developers', 'Publishers', # Strings
    'Categories', 'Genres', 'Tags' # Objects - Lists of Strings
]
```

```python
all_categories = set()
all_genres = set()
all_tags = set()
```

```python
for col in ['Categories', 'Genres', 'Tags']:
    for item in df[col]:
        if isinstance(item, str):
            words = item.split(',')
            if col == 'Categories':
                all_categories.update(words)
            elif col == 'Genres':
                all_genres.update(words)
            elif col == 'Tags':
                all_tags.update(words)
```

```python
print(len(all_categories))
sorted(all_categories)
```

```python
print(len(all_genres))
sorted(all_genres)
```

```python
print(len(all_tags))
sorted(all_tags)
```

```python
all_contents = [item.lower() for item in all_categories.union(all_genres).union(all_tags)]
print(len(all_contents))
sorted(all_contents)
```
