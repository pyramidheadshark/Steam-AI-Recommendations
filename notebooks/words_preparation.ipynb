```python
import os
import json

import pandas as pd
```

```python
project_dir = os.getcwd().split('notebooks')[0]
```

```python
processed_path = os.path.join(project_dir, 'data', 'processed', 'Steam Games 2024.csv')

contents_path = os.path.join(project_dir, 'data', 'raw', 'words.txt')
contents_groups_path = os.path.join(project_dir, 'data', 'raw', 'words_groups.json')

categories_path = os.path.join(project_dir, 'data', 'raw', 'categories.txt')
categories_groups_path = os.path.join(project_dir, 'data', 'raw', 'categories_groups.json')

genres_path = os.path.join(project_dir, 'data', 'raw', 'genres.txt')
genres_groups_path = os.path.join(project_dir, 'data', 'raw', 'genres_groups.json')

tags_path = os.path.join(project_dir, 'data', 'raw', 'tags.txt')
tags_groups_path = os.path.join(project_dir, 'data', 'raw', 'tags_groups.json')
```

```python
df = pd.read_csv(processed_path)
```

```python
columns_to_scale = [
    'AppID', # NOT SCALLABLE
    'Estimated owners', 'Peak CCU', # Integers
    'Positive', 'Negative', 'Score rank', # Integers
    'Achievements', 'Recommendations', # Integers 
    'Developers', 'Publishers', # Strings
    'Categories', 'Genres', 'Tags' # Objects - Lists of Strings
]
```

```python
all_categories = set()
all_genres = set()
all_tags = set()
```

```python
for col in ['Categories', 'Genres', 'Tags']:
    for item in df[col]:
        if isinstance(item, str):
            words = item.split(',')
            if col == 'Categories':
                all_categories.update(words)
            elif col == 'Genres':
                all_genres.update(words)
            elif col == 'Tags':
                all_tags.update(words)
```

```python
print(len(all_categories))
print(len(all_genres))
print(len(all_tags))
```

```python
all_contents = [item.lower() for item in all_categories.union(all_genres).union(all_tags)]
print(len(all_contents))
sorted(all_contents)
```

```python
with open (categories_path, 'w') as f:
    f.write('\n'.join(all_categories))
    f.close()
    
with open (genres_path, 'w') as f:
    f.write('\n'.join(all_genres))
    f.close()
    
with open (tags_path, 'w') as f:
    f.write('\n'.join(all_tags))
    f.close()

with open (contents_path, 'w') as f:
    f.write('\n'.join(all_contents))
    f.close()
```

```python
grouped_categories = []
grouped_genres = []
grouped_tags = []
grouped_contents = []
```

```python
for col in ['Categories', 'Genres', 'Tags']:
    for item in df[col]:
        if isinstance(item, str):
            words = item.split(',')
            if col == 'Categories':
                grouped_categories.append(words)
            elif col == 'Genres':
                grouped_genres.append(words)
            elif col == 'Tags':
                grouped_tags.append(words)
```

```python
for index, row in df.iterrows():
    if isinstance(row['Categories'], str): categories = row['Categories'].split(',')
    if isinstance(row['Genres'], str): genres = row['Genres'].split(',')
    if isinstance(row['Tags'], str): tags = row['Tags'].split(',')
    grouped_contents.append(categories + genres + tags)
```

```python
print(len(grouped_categories))
print(len(grouped_genres))
print(len(grouped_tags))
```

```python
print(len(grouped_contents))
sorted(grouped_contents)
```

```python
with open(categories_groups_path, 'w') as f:
    json.dump(grouped_categories, f, indent=4)
    f.close()
    
with open(genres_groups_path, 'w') as f:
    json.dump(grouped_genres, f, indent=4)
    f.close()
    
with open(tags_groups_path, 'w') as f:
    json.dump(grouped_tags, f, indent=4)
    f.close()
    
with open(contents_groups_path, 'w') as f:
    json.dump(grouped_contents, f, indent=4)
    f.close()
```
